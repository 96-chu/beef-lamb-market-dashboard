# market_quarterly_summary

```python
import pandas as pd

exports_q = pd.read_csv("../data/processed/exports_quarterly_2025_07_to_2025_12.csv")
production = pd.read_csv("../data/processed/production_clean_latest_2025_07_to_2025_12.csv")

exports_summary = (
    exports_q.groupby(["quarter", "product"], as_index=False)["tonnes"]
    .sum()
    .rename(columns={"tonnes": "exports_tonnes"})
)

production_summary = (
    production[
        (production["metric_group"] == "production") &
        (production["unit"] == "tonnes") &
        (production["state"] == "Australia")
    ]
    .groupby(["quarter", "product"], as_index=False)["value"]
    .sum()
    .rename(columns={"value": "production_tonnes"})
)

slaughter_summary = (
    production[
        (production["metric_group"] == "slaughter") &
        (production["unit"] == "head") &
        (production["state"] == "Australia")
    ]
    .groupby(["quarter", "product"], as_index=False)["value"]
    .sum()
    .rename(columns={"value": "slaughter_head"})
)

market_summary = (
    exports_summary
    .merge(production_summary, on=["quarter", "product"], how="outer")
    .merge(slaughter_summary, on=["quarter", "product"], how="outer")
    .sort_values(["quarter", "product"])
)

market_summary.to_csv(
    "../data/processed/market_quarterly_summary_2025_07_to_2025_12.csv",
    index=False
)

print(market_summary)
```
